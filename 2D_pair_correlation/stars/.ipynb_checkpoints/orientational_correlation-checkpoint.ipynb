```python
import pylab
import numpy as np
import numpy.random
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
def orient_corref(M, N, NP, L, n, binr, name):

    r = np.linspace(0.1111111111, L/2, binr)
    results = []
    
    for run in range(0,10):

        dist = {}
        for i in range(len(r)):
            dist[i] = []

        for i in range(M,N):
            data = pylab.loadtxt('./' + str(run) + '_' + name + '_' + str(round(L,2)) + '/traj_MC_' + name + '_' + str(round(L,2)) + '_' + str(i) + '.dat')
            x = data[:,0]
            y = data[:,1]
            theta = data[:,2]

            for j in range(0,NP):
                for k in range(j,NP):
                    dx = abs(x[j]-x[k])
                    dy = abs(y[j]-y[k])

                    #periodic boundary condition
                    if dx>L/2: dx = L-dx
                    if dy>L/2: dy = L-dy

                    d = (dx**2 + dy**2)**0.5

                    t1 = theta[j]
                    if t1 >= np.pi:
                        t1 = t1 - np.pi

                    t2 = theta[k]
                    if t2 >= np.pi:
                        t2 = t2 - np.pi

                    for l in range(0, len(r)-1):
                        if d > r[l] and d < r[l+1]:
                            dist[l+1].append(t1*t2)

        res = []

        for i in range(0,len(dist)-1):
            res.append(np.average(dist[i]))

        for i in range(len(res)):
            norm = 2*np.pi*((r[i+1]))*(r[1] - r[0])/(n*(N-M))
            res[i] = res[i]/norm

        res = res
        
        results.append(res)
    
    std = np.std(results, axis=0)
    results = np.average(results, axis=0)
    
#     results = results/min(results[1::])
    return results, std, r
```

```python
num_disk = 31
Area_frac = 0.20
D = 0.11111111111
phi = 4*Area_frac/(num_disk*np.pi*D*D)

M = 1
N = 100
NP = 64
L = (NP/phi)**0.5
n = N/(L**2)
binr = 25

name = 'NORMAL'
results, std, r = orient_corref(M, N, NP, L, n, binr, name)
plt.errorbar(r[0:-1], results, yerr=std, ecolor='red', elinewidth=1, capsize=5, color='red', label='Explicit Distance Calculation')

name = 'ML_Grad_Boost'
results, std, r = orient_corref(M, N, NP, L, n, binr, name)
plt.errorbar(r[0:-1], results, yerr=std, ecolor='black', elinewidth=1, capsize=5, color='black', label='ML Model')

plt.xlabel('r')
plt.ylabel('<theta1, theta2>')
plt.legend()
plt.grid()
plt.show()
```
